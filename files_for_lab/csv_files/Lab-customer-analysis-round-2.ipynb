```python
import pandas as pd
dataset = pd.read_csv('marketing_customer_analysis.csv')
dataset
```

```python
dataset.shape
```

```python
col_std = []
for col in dataset.columns:
    col_std.append(col.lower())
dataset.columns = col_std
dataset.columns = dataset.columns.str.replace(' ', '_')
dataset
```

```python
dataset.dtypes
```

```python
dataset.isna().sum()
```

```python
#Checking if there is any pattern in the data which could help replace the NaN
print(dataset['state'].value_counts())
print(dataset['response'].value_counts())
print(dataset['months_since_last_claim'].value_counts())       
print(dataset['number_of_open_complaints'].value_counts())       
print(dataset['vehicle_class'].value_counts())       
print(dataset['vehicle_size'].value_counts())
print(dataset['vehicle_type'].value_counts())
```

```python
clean_data = dataset.drop(["unnamed:_0","vehicle_type"], axis=1) #Dropping unnecessary columns
clean_data = clean_data.dropna() #Dropping all the ~600 NaN entries in other columns
clean_data
```

```python
clean_data.isna().sum() #checking that it worked
```

```python
clean_data['month'] = pd.DatetimeIndex(clean_data['effective_to_date']).month #adding column and extracting the number which indicates the month
clean_data
```

```python
clean_data['month'].value_counts() #checking the months which are present = already contains data only for january and february
```
