```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('marketing_customer_analysis.csv')
dataset
```

```python
dataset.shape
```

```python
col_std = []
for col in dataset.columns:
    col_std.append(col.lower())
dataset.columns = col_std
dataset.columns = dataset.columns.str.replace(' ', '_')
dataset
```

```python
dataset.dtypes
```

```python
dataset.isna().sum()
```

```python
#Checking if there is any pattern in the data which could help replace the NaN
print(dataset['state'].value_counts())
print(dataset['response'].value_counts())
print(dataset['months_since_last_claim'].value_counts())       
print(dataset['number_of_open_complaints'].value_counts())       
print(dataset['vehicle_class'].value_counts())       
print(dataset['vehicle_size'].value_counts())
print(dataset['vehicle_type'].value_counts())
```

```python
clean_data = dataset.drop(["unnamed:_0","vehicle_type"], axis=1) #Dropping unnecessary columns
clean_data = clean_data.dropna() #Dropping all the ~600 NaN entries in other columns
clean_data
```

```python
clean_data.isna().sum() #checking that it worked
```

```python
clean_data['month'] = pd.DatetimeIndex(clean_data['effective_to_date']).month #adding column and extracting the number which indicates the month
clean_data
```

```python
clean_data['month'].value_counts() #checking the months which are present = already contains data only for january and february
```

```python
clean_data["effective_to_date"] = pd.to_datetime(clean_data["effective_to_date"])      #create new column to store only month and year (fewer different data points will be better for visualization)                
num = clean_data.select_dtypes(include=np.number)
num.describe(include='all').T
```

```python
cat = clean_data.select_dtypes(include=object)
cat.describe(include='all').T
```

```python
for column in num.columns:
    sns.displot(num[column])
    plt.show()
```

```python
for column in cat.columns:
    plt.hist(cat[column])
    plt.show()
```

```python
#Normalising the numerical data 
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

transformer = MinMaxScaler().fit(num)
num_normalized = transformer.transform(num)
print(num_normalized.shape)
num_normalized=pd.DataFrame(num_normalized, columns=num.columns)
```

```python
num_normalized.describe().T
```

```python
for column in num_normalized.columns:
    sns.displot(num_normalized[column])
    plt.show()
```

```python
correlations = num.corr()
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(correlations, annot=True)
plt.show()
```
